```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
# Read in the City Weather Data.csv
city_weather_data = pd.read_csv("City Weather Data.csv")
city_weather_data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps with API key
gmaps.configure(api_key=g_key)
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations=city_weather_data[["Lat", "Lon"]].astype(float)
# Convert Poverty Rate to float and store
humidity=city_weather_data["Humidity"].astype(float)
```

```python
# Create a humidity Heatmap layer
fig=gmaps.figure()

heat_layer=gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=100, point_radius=3)

fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
ideal_temps=city_weather_data[(city_weather_data["Max Temp"]<75)&(city_weather_data["Max Temp"]>70)]
ideal_temps.head()
```

```python
ideal_wind=ideal_temps[(ideal_temps["Wind Speed"]<=12)]
ideal_wind.head()
```

```python
ideal_clouds=ideal_wind[(ideal_wind["Cloudiness"]<=25)& (ideal_wind["Cloudiness"]>=15)]
ideal_clouds
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df=ideal_clouds
hotel_df["Hotel Name"]=""
hotel_df
```

```python
# Loop through the hotel_df and get the hotel closest to each city
target_radius=5000
target_search="hotel"
target_type="lodging"
params={"key":g_key,
        "radius": target_radius,
        #"rankby": "distance",
        "keyword": target_search,
        "type": target_type}

# use iterrows to iterate through pandas dataframe
for index, row in hotel_df.iterrows():
    # get lat and lon from df
    latitude = row["Lat"]
    longitude=row["Lon"]
    
    # update location key value in params dict
    params["location"] = f"{latitude},{longitude}"
    
    # use the search term "international airport" and the appropriate lat/lng
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    # assemble url and make API request & convert to JSON
    response=requests.get(base_url, params=params).json()
    
    # print the response
    #print(index, response["results"][0]["name"])
    
    try:
        hotel_df.loc[index, "Hotel Name"]=response["results"][0]["name"]
        print("Hotel Found!")
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
    
```

```python
# Visualize to confirm hotel data appears
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lon"]]
```

```python
# Add marker layer ontop of heat map
hotel_layer = gmaps.marker_layer(
    locations, info_box_content=[info_box_template.format(**row) for index, row in hotel_df.iterrows()])

# Display figure
fig=gmaps.figure()
fig.add_layer(hotel_layer)
fig.add_layer(heat_layer)
fig
```
